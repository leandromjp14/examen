```python
# crea un menu de python con lo siguiente 1 comprar entradas,2mostrar ubicaciones disponibles del 1 al 100.3 ver lista de asisente.4 mostrar ganancias totales.5 salir
#cantidad entrada fluctua entre 1 y 3
# el usuario debe selecionar una de las intradas del 1 al 100 y si ya esta ocupada saldra el mensaje d no esta disponible seleciones otra
#
# elija una de las entradas los presios de las entradas son platinum $120.000 asientos del 1 al 20
#gold $80.000 asientos del 21 al 50
#silver $50.000 asientos del 51 al 100
#el programa debe mostrar el mensaje que indique que la operacion se realizo con exito
#el sistema debe mostrar el listado de asistentes por run
#mostrar ganancias totales

ubicaciones = ["Platinum"] * 20 + ["Gold"] * 30 + ["Silver"] * 50
ubicaciones_vendidas = [False] * len(ubicaciones)
precios = {"Platinum": 120000, "Gold": 80000, "Silver": 50000}
asistentes = []




input("vienvenido a Creativos.cl(presione enter para continuar)")

def comprar_entradas():
    cantidad = int(input("Ingrese la cantidad de entradas a comprar entre(1-3):"))
    while cantidad < 1 or cantidad > 3:
        print("Cantidad erronea. Por favor, ingrese un valor entre 1 y 3.")
        cantidad = int(input("Ingrese la cantidad de entradas a comprar (1-3):"))




    print("Ubicaciones disponibles:")
    for i in range(len(ubicaciones)):
        if not ubicaciones_vendidas[i]:
            print(f"{i + 1}. {ubicaciones[i]}")
        else:
            print(f"{i + 1}. [Vendido]")



    for _ in range(cantidad):
        seleccion = int(input("Seleccione una ubicación: ")) - 1
        if ubicaciones_vendidas[seleccion]:
            print("La ubicación seleccionada no está disponible.")
        else:

            run = input("Ingrese el RUN(sin guiones ni dígito verificador): ")
            asistentes.append({"run": run, "ubicacion": ubicaciones[seleccion]})
            ubicaciones_vendidas[seleccion] = True




    print("!Operación realizada con exito.!")






def mostrar_ubicaciones_disponibles():
    print("Ubicaciones disponibles:", [i for i in range(1, 101)])




def ver_lista_asistentes():
    print("lista de asistentes.")




def mostrar_ganancias_totales():
    total_ganancias = sum(precios[ubicacion["ubicacion"]] for ubicacion in asistentes)
    print(f"Ganancias totales: {total_ganancias}")




while True:
    print("----- MENU -----")
    print("1. Comprar entradas")
    print("2. Mostrar ubicaciones disponibles 1-100")
    print("3. Ver lista de asistentes")
    print("4. Mostrar ganancias totales")
    print("5. Salir")


    opcion = input("elija una opción (1-5): ")





    if opcion == "1":
        comprar_entradas()
    elif opcion == "2":
        mostrar_ubicaciones_disponibles()
    elif opcion == "3":
        ver_lista_asistentes()
    elif opcion == "4":
        mostrar_ganancias_totales()
    elif opcion == "5":
        print("¡Hasta luego!")
        break
    else:
        print("Opción inválida. Por favor, seleccione una opción válida.")
```
